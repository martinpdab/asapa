```python
import numpy as np
import pandas as pd
```

```python
#memakai dataset
URL = 'https://raw.githubusercontent.com/martinpdab/tugasscdl/main/dataset.csv' 
dataframe = pd.read_csv(URL)
dataframe
```

```python
dataframe.head(n=5)
dataframe.tail(n=6)

dataframe.info()
```

```python
data = dataframe[['TE', 'HU', 'RF', 'WS', 'PD', 'ABJ', 'LAG0', 'LAG1', 'LAG2', 'LAG3', 'LAG4', 'LAG5', 'LAG6', 'LAG7', 'LAG8', 'LAG9', 'LAG10', 'LAG11', 'LAG12']]
data.head()
```

```python
#mencari z point
from scipy import stats
import numpy as np
z = np.abs(stats.zscore(data))
print(z)
```

```python
#membuat fungsi mendeteksi outlier
import numpy as np
import pandas as pd
outliers=[]
def detect_outlier(data_1):
    
    threshold=3
    mean_1 = np.mean(data_1)
    std_1 =np.std(data_1)
    
    
    for y in data_1:
        z_score= (y - mean_1)/std_1 
        if np.abs(z_score) > threshold:
            outliers.append(y)
    return outliers
```

```python
for header in data.columns:
  outlier = detect_outlier(data[header])
  print(outlier)
```

```python
#filter dengan z-score
filtered_entries = (z < 3).all(axis=1)

new_df = data[filtered_entries]
new_df
```
